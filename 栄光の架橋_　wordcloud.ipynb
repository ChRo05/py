# 文章の準備

```python
# 文章
text="誰にも見せない泪があった人知れず流した泪があった決して平らな道ではなかったけれど確かに歩んで来た道だあの時想い描いた夢の途中に今も何度も何度もあきらめかけた夢の途中いくつもの日々を越えて辿り着いた今があるだからもう迷わずに進めばいい栄光の架橋へと悔しくて眠れなかった夜があった恐くて震えていた夜があったもう駄目だと全てが嫌になって逃げ出そうとした時も想い出せばこうしてたくさんの支えの中で歩いて来た悲しみや苦しみの先にそれぞれの光があるさあ行こう振り返らず走り出せばいい希望に満ちた空へ誰にも見せない泪があった人知れず流した泪があったいくつもの日々を越えて辿り着いた今があるだからもう迷わずに進めばいい栄光の架橋へと終わらないその旅へと君の心へ続く架橋へと"
text
```

#品質分解

```python
# janomeインストール
!pip install janome

from janome.tokenizer import Tokenizer
jt = Tokenizer()

#分かち書き
text_wakati = list(jt.tokenize(text, wakati=True))
text_wakati
```

```python
# スペース区切りの文章にする
text_wakati_space= " ".join(map(str,text_wakati))
text_wakati_space
```

# WordCloud

```python
# WordCloudをインポート
from wordcloud import wordcloud
```

```python
# 日本語フォントのインストール
!apt-get -y install fonts-ipafont-gothic
# 日本語フォント指定
fpath = '/usr/share/fonts/truetype/fonts-japanese-gothic.ttf' 
```

```python
# mask画像を読み込み
from PIL import Image
import numpy as np
hira_mask = np.array(Image.open("/content/hira.png"))

#　WordCloudのパラメータ設定
wordcloud = WordCloud(
    font_path=fpath, # 日本フォントを設定したパス 
    prefer_horizontal=1,　# 横書き
    width=100,
    height=50,
    background_color="black", #背景を白にする
    collocations=False,
    #STOPWORDS=set(["なか","なっ","けれど"]),  #非表示ワードを設定 
    #max_words=50 , #最大単語数
    mask = hira_mask
    )

# WordCloud画像を作成
wordcloud_img = wordcloud.generate(text_wakati_space)

# 描画
import matplotlib.pyplot as plt

#show
plt.figure(figsize=(6,6), dpi=200)
plt.imshow(wordcloud_img,interpolation="bilinear")
plt.axis("off") # メモリを非表示
plt.show()
```
